```python
"""IDENTIFYING OUTLIERS IN THE BOX PLOT OF TOTAL AMOUNT SPENT
OUTLIERS ARE USED TO CALCULATE VALUES THAT ARE WITHIN 3 STANDARD DEVIATIONS
 '+/- 1.5*INTERQUARTILE_RANGE ' IS ROUGHLY 3 STANDARD DEVIATIONS
 OUTLIERS ARE USED TO NORMALIZE THE DISTRIBUTION OF A DATASET
"""
```

```python
first_quartile = df['MntTotal'].quantile(0.25)
third_quartile = df['MntTotal'].quantile(0.75)
Inter_quartile_range = third_quartile - first_quartile
#determining the lower and upper class boundary
lower_class_boundary = first_quartile - 1.5 * Inter_quartile_range
upper_class_boundary = third_quartile + 1.5 * Inter_quartile_range
#values greater than the upper or lower class boundaries
outliers = df[(df['MntTotal'] < lower_class_boundary) | (df['MntTotal'] > upper_class_boundary)]
outliers.head()
```

```python
"""feature engineering 
-first feature describing the higher educational Status of the customers """ 
#GETTING ALL THE COLUMNS READY
df.columns
```

```python
#Higher Educational status description
def higher_educational_status_description(df):
    if df["education_Graduation"]==1:
        return 'Graduate'
    elif df["education_Master"]==1:
         return 'Master'
    elif df["education_PhD"]==1:
        return 'Professor'
    else:
        return 'Unknown Higher Education Status'
#Applying the marital status of new customers to a nw column 
df["Higher_Educational_Status"]= df.apply(higher_educational_status_description,axis=1)
    
```

```python
#Total Amount Spent by higher educational status
Total_Amount_Spent_by_higher_educational_status = df.groupby("Higher_Educational_Status")["MntTotal"].sum().to_frame()
ranked_Total_Amount_Spent_by_higher_educational_status=Total_Amount_Spent_by_higher_educational_status.sort_values("MntTotal",ascending= False)
ranked_Total_Amount_Spent_by_higher_educational_status
```

```python
ranked_Total_Amount_Spent_by_higher_educational_status.plot(kind="bar",xlabel="Higher_Educational_Status",ylabel="MntTotal")
```

```python
#Average Income Received by Educational Status
Average_Income_Received_by_higher_educational_status = df.groupby("Higher_Educational_Status")["Income"].mean().to_frame()
ranked_Average_Income_Received_by_higher_educational_status=Average_Income_Received_by_higher_educational_status.sort_values("Income",ascending= False)
ranked_Average_Income_Received_by_higher_educational_status
```

```python
ranked_Average_Income_Received_by_higher_educational_status.plot(kind="bar",xlabel="Higher_Educational_Status",ylabel="Income")
```

```python
#Professorship status 
def check_professorship(df):
    if df["Higher_Educational_Status"]== "Professor":
        return "A professor"
    else:
        return "Not a Professor"
df["phD Status"]= df.apply(check_professorship,axis=1)
df["phD Status"]
```
